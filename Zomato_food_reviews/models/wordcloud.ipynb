```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import os
```

```python
#setting path to directory
csv_directory = "C:/Users/annar/OneDrive/Desktop/Visual Studio/GitHub/Projects2/Insighter/Zomato_food_reviews/Data"
os.chdir(csv_directory)
```

```python
data = pd.read_csv("./csv_files/100N.csv")
data.head()
```

```python
text = " ".join(word.split()[0] for word in data.Reviews)
text
```

```python
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
wc=WordCloud(stopwords=stop_words, collocations=False, max_font_size=55, max_words=100, background_color="black")
wc.generate(' '.join(data.Reviews))
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```
